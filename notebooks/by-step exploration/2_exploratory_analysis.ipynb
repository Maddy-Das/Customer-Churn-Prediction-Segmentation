```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Reload clean dataset for EDA
df = pd.read_csv('../../data/customer_data.csv')
```

```python
# Visualize churn distribution
sns.countplot(data=df, x='Churn')
plt.title('Customer Churn Distribution')
plt.show()
```

```python
# Check churn rate across contract types
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn Rate by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap for feature correlation
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of MonthlyCharges by Churn
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='cool')
plt.title('Monthly Charges by Churn')
plt.tight_layout()
plt.show()
```

```python
selected_cols = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
sns.pairplot(df[selected_cols], hue='Churn', palette='husl')
plt.show()
```
